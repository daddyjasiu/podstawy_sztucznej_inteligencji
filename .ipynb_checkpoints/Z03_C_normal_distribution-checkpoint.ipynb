# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
import pandas as pd
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
n = [0, 0]
cov = [[1,0],[0,1]]
x_axis = np.linspace(-5,5,500)
y_axis = np.linspace(-5,5,500)
X, Y = np.meshgrid(x_axis,y_axis)
points = multivariate_normal(n, cov)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, points.pdf(pos))
plt.show()
```

```python
n = [0, 0]
cov = [[2,0],[0,1]]
x_axis = np.linspace(-5,5,500)
y_axis = np.linspace(-5,5,500)
X, Y = np.meshgrid(x_axis,y_axis)
points = multivariate_normal(n, cov)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, points.pdf(pos))
plt.show()
```

# Zad 
Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
n = [0, 0]
cov = [[1, 0], [0, 1]]
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,5))
x_axis = np.linspace(-3,3,500)
y_axis = np.linspace(-3,3,500)
X, Y = np.meshgrid(x_axis,y_axis)
points = multivariate_normal(n, cov)
pos = np.dstack((X,Y))
ax.contour(X, Y, points.pdf(pos))
```

# Zad 
Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

i narysuj odpowiedni wykres

```python
n=[0,0]
cov1 = [[1,0],[0,1]]
cov2 = [[2,0],[0,1]]
cov3 = [[1,0],[0,2]]

a, b = np.random.multivariate_normal(n, cov1, 5000).T
c, d = np.random.multivariate_normal(n, cov2, 5000).T
e, f = np.random.multivariate_normal(n, cov3, 5000).T

plt.plot(a, b, '.')
plt.show()
plt.plot(c, d, '.')
plt.show()

plt.plot(e, f, '.')
plt.show()
```
