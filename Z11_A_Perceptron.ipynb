````python
import numpy as np
import sklearn.datasets
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
````

# Zadanie 1
Za pomoca funkcji
```python
sklearn.datasets.make_moons
```
wygeneruj 4 zbiory danych z parametrami:

* n_samples=1000, noise=.05
* n_samples=1000, noise=.3
* n_samples=200, noise=.05
* n_samples=200, noise=.3

````python
#low noise, plenty of samples, should be easy
X, y = sklearn.datasets.make_moons(n_samples=1000, noise=.05)
X2, y2 = sklearn.datasets.make_moons(n_samples=1000, noise=.3)
X3, y3 = sklearn.datasets.make_moons(n_samples=200, noise=.05)
X4, y4 = sklearn.datasets.make_moons(n_samples=200, noise=.3)
````

````python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.show()
plt.scatter(X3[:,0],X3[:,1],c=y3)
plt.show()
plt.scatter(X4[:,0],X4[:,1],c=y4)

plt.axis('equal')
plt.show()
````

````python
iris = load_iris()
X = iris.data[:, (2, 3)]  # petal length, petal width
y = (iris.target == 0).astype(np.int)

per_clf = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X, y)

per_clf2 = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X2, y2)

per_clf3 = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X3, y3)

per_clf4 = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X4, y4)

y_pred = per_clf.predict([[2, 0.5]])
````

````python
from mlxtend.plotting import plot_decision_regions
````

````python
from sklearn.linear_model import Perceptron
per_clf = Perceptron(random_state=42)
per_clf.fit(X, y)

per_clf2 = Perceptron(random_state=42)
per_clf2.fit(X2, y2)

per_clf3 = Perceptron(random_state=42)
per_clf3.fit(X3, y3)

per_clf4 = Perceptron(random_state=42)
per_clf4.fit(X4, y4)
````

````python
plot_decision_regions(X, y, clf=per_clf)
plt.show()
plot_decision_regions(X2, y2, clf=per_clf2)
plt.show()
plot_decision_regions(X3, y3, clf=per_clf3)
plt.show()
plot_decision_regions(X4, y4, clf=per_clf4)
plt.show()
````

# Zad.
Naucz perceptron na danych iris i wylicz F1 score.

````python
iris = load_iris()
X = iris.data[:, (2, 3)] 
y = (iris.target == 0).astype(np.int) 

per_clf = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
per_clf.fit(X, y)

sklearn.metrics.f1_score(y, per_clf.predict(X))
````
