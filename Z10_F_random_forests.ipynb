````python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
````

# Ciekawostka - Extra-Trees 


* Gdy budujemy kolejne drzewa  w losowym lesie, w kadym wle rozpatrywany jest losowy podzbir elementw do podziau. 

* Moliwe jest stworzenie drzewa jeszcze bardziej losowego, wykorzystujc losowe podziay dla kadej cechy, zamiast szuka najlepszych moliwych podziau (jak w przypadku zwykych drzew decyzyjnych).

* Taki las losowy nazywa si **Extremely Randomized Trees** (lub **Extra-Trees** w skrcie). 

* Moesz utworzy klasyfikator **Extremely Randomized Trees** za pomoc klasy **ExtraTreesClassifier** z Scikit-Learn. 

* Jego API jest identyczne z klas RandomForestClassifier.
